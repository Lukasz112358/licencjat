```julia
using Random
using Primes
using Plots
function densityOfPrimes(multiplier, powerOf2,n)
    ans = 0
    for i in 1:n
        p = BigInt(rand(UInt64)%BigInt(multiplier))*powerOf2+1
        if isprime(p)
            ans+=1
        end
    end
    return ans/n
end
```

```julia
densityOfPrimes(10,1024*1024,20000)
```

```julia
function density1(x)
    densityOfPrimes(x,1,100000)*log(x)
end
function density2(x)
    densityOfPrimes(x,2,100000)*log(x)
end
function density4(x)
    densityOfPrimes(x,4,100000)*log(x)
end
function density8(x)
    densityOfPrimes(x,8,100000)*log(x)
end
function density16(x)
    densityOfPrimes(x,16,100000)*log(x)
end
function density64(x)
    densityOfPrimes(x,64,100000)*log(x)
end
function density256(x)
    densityOfPrimes(x,256,100000)*log(x)
end
function density1024(x)
    densityOfPrimes(x,1024,100000)*log(x)
end
function densityMilion(x)
    densityOfPrimes(x,1024*1024,100000)*log(x)
end
function densityBilion(x)
    densityOfPrimes(x,1024*1024*1024,100000)*log(x)
end
function densityQuadrilion(x)
    densityOfPrimes(x,1024*1024*1024*1024*1024,100000)*log(x)
end
```

```julia
1024*1024*1024*1024*1024
```

```julia
x = range(1,100000000000000,step=1000000000000)
```

```julia
y1=density1.(x)
y2=density2.(x)
y4=density4.(x)
y8=density8.(x)
y16=density16.(x)
y64=density64.(x)
y256=density256.(x)
y1024=density1024.(x)
yMilion=densityMilion.(x)
yBilion=densityBilion.(x)
yQuadrilion=densityQuadrilion.(x)
```

```julia
plot()
plot!(x,y1,label="1")
plot!(x,y2,label="2")
plot!(x,y4,label="4")
plot!(x,y8,label="8")
plot!(x,y16,label="16")
plot!(x,y64,label="64")
plot!(x,y256,label="256")
plot!(x,y1024,label="1024")
plot!(x,yMilion,label="1048576")
plot!(x,yBilion,label="1073741824")
plot!(x,yQuadrilion,label="1125899906842624")
```
